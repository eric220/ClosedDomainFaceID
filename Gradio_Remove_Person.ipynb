```python
import pickle
import gradio as gr
```

```python
def get_names_encodings(file_path):
    with open(file_path, 'rb') as handle:
        return pickle.load(handle)
```

```python
def save_data(data, filename):
    with open(filename, 'wb') as handle:
        pickle.dump(data, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
names = get_names_encodings('data/processed/lfw_names.pickle')
encodings = get_names_encodings('data/processed/lfw_encodings2.pickle')
```

```python
assert len(names) == len(encodings)
```

```python
master_name = ''
def search(name):
    global master_name
    msg = '{} was {} found in data' .format(name, '')
    if name not in names:
        msg = 'Name {} was {} found in data' .format(name, 'not')
    master_name = name
    return msg
    
def remove_names_encodings(name_to_remove):
    if (name_to_remove in names):
        for i, n in enumerate(names):
            if n == name_to_remove:
                names.pop(i)
                encodings.pop(i)
                remove_names_encodings(name_to_remove)
    else:
        save_data(names, 'data/processed/lfw_names.pickle')
        save_data(encodings, 'data/processed/lfw_encodings2.pickle')
        
def remove_name():
    remove_names_encodings(master_name)
    return '{} removed from data'.format(master_name)

search_name = gr.Interface(fn = search, inputs = 'text', outputs = 'text')
remove_person = gr.Interface(fn = remove_name, inputs = None, outputs = 'text')
demo = gr.TabbedInterface([search_name, remove_person], [ "Search Name", "Remove Name"])

if __name__ == "__main__":
    demo.launch()
```
